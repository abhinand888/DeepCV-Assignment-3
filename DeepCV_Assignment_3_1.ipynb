Q2.Implement CNN for classifying an apple's image kept in any 1 of 4 quadrants.

```python
import os
import random
from PIL import Image


apple_image = Image.open("/content/red-apple-isolated-clipping-path-19130134.jpg")

img_width, img_height = apple_image.width*4, apple_image.height*4

quadrants = ['top_left', 'top_right', 'bottom_left', 'bottom_right']


if not os.path.exists('apple_quadrants'):
    os.makedirs('apple_quadrants')


for quadrant in quadrants:
    for i in range(100):
        
        new_img = Image.new('RGB', (img_width, img_height), color='white')

       
        x = random.randint(0, img_width//2 - apple_image.width)
        y = random.randint(0, img_height//2 - apple_image.height)

        
        if quadrant == 'top_left':
            new_img.paste(apple_image, (x, y))
        elif quadrant == 'top_right':
            new_img.paste(apple_image, (x + img_width//2, y))
        elif quadrant == 'bottom_left':
            new_img.paste(apple_image, (x, y + img_height//2))
        elif quadrant == 'bottom_right':
            new_img.paste(apple_image, (x + img_width//2, y + img_height//2))

       
        filename = f'{quadrant}_{i}.jpg'
        filepath = os.path.join('apple_quadrants', filename)
        new_img.save(filepath)
        print(f'Saved {filepath}')
```

```python
import os
import random
import numpy as np
from PIL import Image
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras import layers


data_dir = 'apple_quadrants'
quadrants = ['top_left', 'top_right', 'bottom_left', 'bottom_right']
data = []
labels = []
for quadrant in quadrants:
    for i in range(100):
        filename = f'{quadrant}_{i}.jpg'
        filepath = os.path.join(data_dir, filename)
        img = Image.open(filepath)
        img = img.resize((200, 200)) # Resize the images to a fixed size
        img_arr = np.array(img) / 255.0 # Convert the image to a numpy array and normalize
        data.append(img_arr)
        labels.append(quadrants.index(quadrant))


data = np.array(data)
labels = np.array(labels)


x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, random_state=42)


model = keras.Sequential(
    [
        layers.Conv2D(32, (3, 3), activation="relu", input_shape=(200, 200, 3)),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, (3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(128, (3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dense(64, activation="relu"),
        layers.Dense(4, activation="softmax"),
    ]
)

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])


model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_acc}")
```
